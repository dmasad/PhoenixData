# Downloading the data

The first thing we need to do is download the data. Currently, the [Phoenix data](http://phoenixdata.org/data) is stored as zip files, one per day of data collected. 

A quick glace at the page itself shows the pattern of file names.

```python
BASE = "https://s3.amazonaws.com/oeda/data/" # This is the URL for the directory
DATA = "/Users/dmasad/Data/Phoenix/" # Wherever you want to download the data to.
```

Currently, the earliest date of data is June 20th, 2014. To download the data, we'll loop over every date between that and yesterday (February 17th, 2015 for me) and try to download the corresponding file.

It's not too hard to download files using only Python's standard library, but the [requests](http://docs.python-requests.org/en/latest/) package makes it just a little bit easier. Once it's downloaded into memory, we'll use Python's built-in [zipfile](https://docs.python.org/3.4/library/zipfile.html) library to extract it directly. This keeps us from cluttering our computer with too many extra zip files.

**caution:** using the [*extractall*](https://docs.python.org/3.4/library/zipfile.html#zipfile.ZipFile.extractall) method to extract zip files in place isn't generally very secure. In this case, I trust the OEDA folks and am willing to risk it. But in general, this isn't the best practice.

```python
start_date = dt.datetime(2014, 6, 20).date()
today = dt.datetime.today().date()
date = start_date

# Loop over all the dates
while date != today:
    date_str = date.strftime("%Y%m%d") # Convert date to target format
    file_name = "events.full." + date_str + ".txt.zip"
    # Download the file
    r = requests.get(BASE + file_name, stream=True) # Stream the file into memory
    if r.status_code == 200: # Make sure the request status is OK
        z = ZipFile(BytesIO(r.content)) # Handle downloaded object as a zip file
        z.extractall(DATA) # Extract it to your desired data directory
        # print(file_name) # If you want to log every download, you can print it.
    else:
        print("Error with {}, status_code: {}".format(date_str, r.status_code))
    date += dt.timedelta(1)
```

You'll get some 403 errors like I did for days when there seems to be data missing.  You may also want to manually check the files, to make sure that they all extracted properly. When I ran this code, the **events.full.20150106.txt** had a trailing *.zip* added to the end, despite being a correctly-formatted text file. I just went ahead and fixed the extension by hand.

## Read in all data

You probably don't want to look at just one day of data. If the data were much larger, it would get impractical to hold it all in memory -- but at the moment it's still small enough to read into a DataFrame. To do it, we'll iterate over all the data files, read each into a DataFrame, and then concatenate them together.

This code assumes that the Phoenix data files are the only .txt files in your DATA directory. If that isn't the case, adjust as needed.

```python
all_data = []
for f in os.listdir(DATA): # Iterate over all files
    if f[-3:] != "txt":  # Skip non-text files.
        continue
    df = pd.read_csv(DATA + f, sep='\t', names=col_names)
    all_data.append(df)

data = pd.concat(all_data)
```

Another good sanity check, which [Andy Halterman](https://twitter.com/ahalterman) does in his [example R code](https://s3.amazonaws.com/oeda/docs/code_example.pdf) is to get a count of how often each actor appears as the source or target of an event.

```python
actors_source = data.SourceActorFull.value_counts()
actors_target = data.TargetActorFull.value_counts()
actor_counts = pd.DataFrame({"SourceFreq": actors_source,
                             "TargetFreq": actors_target})
actor_counts.fillna(0, inplace=True)
actor_counts["Total"] = actor_counts.SourceFreq + actor_counts.TargetFreq
```

### Counting events by PentaClass

Counting events by broad category is a good way of turning discrete events into a rough time series. Phoenix expands the venerable QuadClass event categorization (Material Cooperation, Verbal Cooperation, Verbal Conflict, Material Conflict) by adding a category 0, recognizing that not every event involves cooperation or conflict. 

Let's start with all Israel-originating events and get a count by category, by day. Here's the Pythonic way of doing it.

```python
ilpalcon = ilpalcon[ilpalcon.Year >= 2014] # Drop out-of-range events
```

```python
il_initiated = ilpalcon[ilpalcon.SourceActorEntity=="ISR"]
grouped = il_initiated.groupby(["Date", "PentaClass"])
il_initiated = grouped.aggregate(len)["EventID"].reset_index()
```

```python
# Reshape the data with PentaClass as columns
il_initiated = il_initiated.pivot(index="Date", columns="PentaClass", 
                                  values="EventID")
il_initiated.fillna(0, inplace=True)
```

```python
# And finally plot:
il_initiated.plot(figsize=(12,6), title="Israel-Initiated Actions")
```

#### Non Pythonic way

Pandas's syntax makes it natural to want to chain together multiple methods to manipulate our data. This is more compact than the way above, though isn't considered very Pythonic. To avoid writing one long, unreadable line of code, the  the \ character lets you chain multiple lines together, like this:

```python
il_initiateed = ilpalcon[ilpalcon.SourceActorEntity=="ISR"] \
                    .groupby(["Date", "PentaClass"]) \
                    .aggregate(len)["EventID"].reset_index() \
                    .pivot(index="Date", columns="PentaClass", values="EventID") \
                    .fillna(0)
```

```python
il_initiated.plot(figsize=(12,6), title="Israel-Initiated Actions")
```

These results look fairly good: we see a spike in events (and especially Class 4, Material Conflict events) during the [2014 Gaza Conflict](http://en.wikipedia.org/wiki/2014_Israel%E2%80%93Gaza_conflict), which simmers down later. 

A couple other things to notice: 

* There are still some events in there from outside the range of days that were parsed. 

* During the conflict there's a spike in *all* event counts, not just conflict events. This is likely at least in part because there's simply more media attention being paid during this time, which in turn generates more events. 

# Mapping

Let's get a sense for Phoenix's overall coverage by putting it on a map. Doing this right really requires the [Basemap toolkit](http://matplotlib.org/basemap/) (see for example [this](http://nbviewer.ipython.org/github/dmasad/GDELT_Intro/blob/master/GDELT_Mapping.ipynb)). Unfortunately, installing Basemap on OSX Mavericks inside of a virtual environment is outside the scope of this tutorial, or my energy for today. 

So I'm going to do it wrong. Below, I group the events based on their geocoded Lat and Long coordinates, counting the number of events at each point. 

To plot them, I'll just use a scatter plot -- no projection, no nothing. Point sizes are based on the log of the number of events that took place at that point. I add 1 to the count of events in order to ensure that points with a single event make it onto the not-really-map too.

```python
points = data.groupby(["Lat", "Lon"]).aggregate(len)["EventID"].reset_index()
```

```python
points.plot(kind='scatter', x='Lon', y="Lat", s=pd.np.log(points.EventID+1)*10,
            alpha=0.4,
            figsize=(12,8), title="The World According to Phoenix")
```

Even with this embarassingly bad visualization, we see something that's recognizably an outline of the world. It even gives us some idea of where there is dense coverage (North America, Europe, and apparently western Africa) and where coverage is sparse (Latin America).

# Issues

Issues are one of Phoenix's new features. Issues are tags attached to events which attempt to add another layer of context. Issues are based on user-provided [keywords](https://github.com/openeventdata/Dictionaries/blob/master/Phoenix.IssueCoding.txt).

Each Issues entry is a miniature table, associating issues with the number of times their keywords appear in the relevant sources. Issues are separated from their counts with commas, and issue-count rows are separated from one another via semi-colons.

```python
data.Issues.value_counts()
```

Let's count up all mentions of each issue, regardless of date or other issue, and see which issues crop up most frequently. 

As far as pandas is concerned, the Issues entries are just plain text. We need to parse them ourselves. Note that many events have no Issues associated with them, which we'll ignore for now.

```python
all_issues = defaultdict(int)
for entry in data.Issues.values:
    if pd.notnull(entry):
        for row in entry.split(";"):
            issue, count = row.split(",")
            all_issues[issue] += int(count)
# Turn the dictionary into a pandas series...
all_issues = pd.Series(all_issues)
all_issues.sort(ascending=False, inplace=True) # ... and sort it.
```

```python
all_issues
```

Perhaps not too surprisingly, terrorism-related issues are at the top of the list. Remember, though, that this is partially an artifact of what keywords are coded into the Phoenix pipeline.

### Issue Network

Are there any interesting patterns in which issues occur together? Let's find out!

[NetworkX](https://networkx.github.io/) is the main Python network analysis library, which I'll use to describe the issue co-occurences as an undirected network.

```python
import networkx as nx
import itertools
```

```python
issue_dyads = defaultdict(int)
for entry in data.Issues.values:
    if pd.notnull(entry):
        entry_issues = []
        for row in entry.split(";"):
            issue, count = row.split(",")
            entry_issues.append(issue)
        # Check all issue pairs:
        for src, dest in itertools.combinations(entry_issues, 2):
            if (dest, src) in issue_dyads:
                issue_dyads[(dest, src)] += 1
            else:
                issue_dyads[(src, dest)] += 1

# Turn the dictionary into a network object
issue_net = nx.Graph()
for dyad, count in issue_dyads.items():
    issue_net.add_edge(dyad[0], dyad[1], weight=count)
```

NetworkX isn't great at visualizing hairball networks, but let's give it a try anyway.

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

pos = nx.spring_layout(issue_net, scale=2.0)
edge_widths = [pd.np.log(edge[2]["weight"]) 
               for edge in issue_net.edges(data=True)]
nx.draw_networkx_edges(issue_net, pos=pos, width=edge_widths, ax=ax, alpha=0.5)
nx.draw_networkx_labels(issue_net, pos=pos, ax=ax, font_size=8, alpha=0.75)
ax.axis('off')
```

More useful here might be building a table of network statistics. NetworkX reports node-level statistics as dictionaries, which are easy to convert to pandas objects:

```python
network_data = pd.DataFrame({"Degree": nx.degree(issue_net),
                             "WeightedDegree": nx.degree(issue_net, weight='weight'),
                             "Closeness": nx.closeness_centrality(issue_net),
                             "Betweenness": nx.betweenness_centrality(issue_net)
                             })
network_data.sort("Degree", ascending=False, inplace=True)
```

```python
network_data.head(10)
```

And we can see the nodes that maximize a few of the values:

```python
print("Max Weighted Degree: {}".format(network_data.WeightedDegree.argmax()))
print("Max Closeness: {}".format(network_data.Closeness.argmax()))
print("Max Betweenness: {}".format(network_data.Betweenness.argmax()))
```

Though named terror groups co-occur with other issues substantially more often, security services co-occur with more issues, and have a much more central position in the issue network. 

We can also look at the distribution of these statistics, for example Degree:

```python
network_data.Degree.plot()
plt.ylabel("Degree")
```

This is the kind of long-tail distribution we expect to see in complex networks, though there may not be quite enough data here to see a real [power law](http://en.wikipedia.org/wiki/Power_law).

(Note that this graph only came out nicely because the DataFrame was pre-sorted. In general, probably not the best practice)
